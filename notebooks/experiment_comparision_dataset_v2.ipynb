```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


def plot_images_side_by_side(img_path1, img_path2, global_title, img1_title, img2_title):
    img1 = mpimg.imread(img_path1)
    img2 = mpimg.imread(img_path2)
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    fig.suptitle(global_title)
    axes[0].imshow(img1)
    axes[0].set_title(img1_title)
    axes[0].axis('off')
    axes[1].imshow(img2)
    axes[1].set_title(img2_title)
    axes[1].axis('off')
    plt.show()
```

```python
plot_images_side_by_side(
    img_path1='./retr_test_backone/glomerulo.png', 
    img_path2='./retr_test_fsl/glomerulo.png', 
    global_title='Glomerulo', 
    img1_title='Backbone', 
    img2_title='FSL'
    )
```
